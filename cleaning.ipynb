```python
def mp(r):
    # Handle ranges like '23-28'
    if '-' in r:
        start, end = r.split('-')
        return (int(start) + int(end)) / 2
    if 'and above' in r:
        return int(r.split()[0]) 

# ds3['Age'].apply(mp)
ds3['Age'] = ds3['Age'].apply(mp)
ds3
```

```python
ds3
```

```python
def pf(freq):

    if '-' in freq:
        lower, upper = map(int, freq.split('-'))
        return (lower + upper) / 2 

    if 'Once last year' in freq:
        return 1
    if 'Just surf and buy nothing' in freq:
        return 0

ds3['Online purchase frequency']=ds3['Online purchase frequency'].apply(pf)
```

```python
ds3
```

```python
def m(r):

    if '-' in r:
        lower, upper = map(int, r.split('-'))
        return (lower + upper) / 2

    if 'under' in r:
        return int(r.split()[1])

ds3['Money spent on these purchases (Yearly)'].apply(m)
```
